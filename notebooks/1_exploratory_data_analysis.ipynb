## Graficos

Distribucion de log_price:

```python
plt.figure(figsize=(10,6))
sns.histplot(df['log_price'], bins=30, kde=True, color='blue')
plt.title("Distribución de log_price")
plt.xlabel("log_price")
plt.ylabel("Frecuencia")
plt.show()

plt.figure(figsize=(8,4))
sns.boxplot(x=df['log_price'], color='lightgreen')
plt.title("Boxplot de log_price")
plt.xlabel("log_price")
plt.show()
```

Distribución de Property Type:

```python
plt.figure(figsize=(10,6))
sns.countplot(x='property_type', data=df, order=df['property_type'].value_counts().index, palette='Set2')
plt.title("Distribución de Property Type")
plt.xlabel("Property Type")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()
```

Distribución de Room Type:

```python
plt.figure(figsize=(10,6))
sns.countplot(x='room_type', data=df, order=df['room_type'].value_counts().index, palette='Set3')
plt.title("Distribución de Room Type")
plt.xlabel("Room Type")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()
```

Correlacion de las variables numericas:

```python
numerical_features = ['log_price', 'accommodates', 'bathrooms', 'number_of_reviews', 
                      'review_scores_rating', 'latitude', 'longitude', 'bedrooms', 'beds']

plt.figure(figsize=(12,8))
sns.heatmap(df[numerical_features].corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de Correlación de Variables Numéricas")
plt.show()
```

log_price vs. review_scores_rating:

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='review_scores_rating', y='log_price', data=df, alpha=0.6, color='purple')
plt.title("log_price vs. Review Scores Rating")
plt.xlabel("Review Scores Rating")
plt.ylabel("log_price")
plt.show()
```

log_price vs. accommodates:

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='accommodates', y='log_price', data=df, alpha=0.6, color='teal')
plt.title("log_price vs. Accommodates")
plt.xlabel("Accommodates")
plt.ylabel("log_price")
plt.show()
```

Boxplot para number_of_reviews:

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['number_of_reviews'], color='orange')
plt.title("Boxplot de Number of Reviews")
plt.xlabel("Number of Reviews")
plt.show()
```

Boxplot para review_scores_rating:

```python
plt.figure(figsize=(8,4))
sns.boxplot(x=df['review_scores_rating'], color='skyblue')
plt.title("Boxplot de Review Scores Rating")
plt.xlabel("Review Scores Rating")
plt.show()
```
